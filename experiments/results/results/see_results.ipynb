```python
import pandas as pd
import numpy as np

df = pd.read_csv("results_full_V3.csv")

result = (
    df.groupby(["dataset", "alpha"], as_index=False)
    .mean(numeric_only=True)
)

ert_cols = ["ERT_levelset", "ERT_likelihood", "ERT_one", "ERT_ot", "ERT_MVCS"]
wsc_cols = ["WSC_levelset", "WSC_likelihood", "WSC_one", "WSC_ot", "WSC_MVCS"]
vol_cols = ["volume_levelset", "volume_likelihood", "volume_one", "volume_ot", "volume_MVCS"]

def underline_best(row):
    styles = [""] * len(row.index)
    
    # underline lowest ERT
    min_ert = row[ert_cols].min()
    for i, col in enumerate(row.index):
        if col in ert_cols and row[col] == min_ert:
            styles[i] = "text-decoration: underline"


    # underline lowest ERT
    min_vol = row[vol_cols].min()
    for i, col in enumerate(row.index):
        if col in vol_cols and row[col] == min_vol:
            styles[i] = "text-decoration: underline"

    # underline highest WSC
    max_wsc = row[wsc_cols].max()
    for i, col in enumerate(row.index):
        if col in wsc_cols and row[col] == max_wsc:
            styles[i] = "text-decoration: underline"

    return styles

styled = result.style.apply(underline_best, axis=1)

styled
```
